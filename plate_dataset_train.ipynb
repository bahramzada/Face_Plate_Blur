```python
# -*- coding: utf-8 -*-
import yaml
from pathlib import Path

ROOT = Path("/workspace")
DATA_YAML = ROOT / "plates_data.yaml"
NAMES = ["plate"]

data = {
    "path": str(ROOT),
    "train": "train/images",
    "val":   "val/images",
    "test":  "test/images",
    "names": {i: n for i, n in enumerate(NAMES)}
}

with open(DATA_YAML, "w", encoding="utf-8") as f:
    yaml.safe_dump(data, f, sort_keys=False, allow_unicode=True)

print("YAML yaradıldı:", DATA_YAML)
```

```python
!yolo detect train \
  model=yolov8l.pt data=/workspace/plates_data.yaml \
  device=0,1 imgsz=1536 epochs=100 batch=16 workers=16 \
  amp=True cache=True multi_scale=False mosaic=0 rect=False \
  optimizer=AdamW lr0=0.0025 lrf=0.1 weight_decay=0.01 cos_lr=True warmup_epochs=3 patience=30 \
  copy_paste=0.0 mixup=0.0 hsv_h=0.015 hsv_s=0.5 hsv_v=0.4 degrees=2.0 translate=0.08 scale=0.20 \
  save_period=10 seed=42 plots=False verbose=False val=False
```
